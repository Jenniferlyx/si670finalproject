```python
import glob
import pandas as pd
results = glob.glob('../data/*.csv')
df = pd.DataFrame()
for path in results:
    if path == "../data/scraped_tweets-posted-11-09-collected-11-16.csv" or path == "../data/scraped_tweets-posted-11-25-collected-12-02.csv":
        continue
    tmp = pd.read_csv(path, lineterminator='\n')
    print(tmp.shape)
    df = pd.concat([df, tmp])
df = df[df['hashtags'] != '#lgbtq#nft']
df.dropna(subset=['hashtags'], inplace=True, axis=0)
df
```

```python
# Functions for preprocessing texts
import nltk
from nltk.metrics.distance import jaccard_distance, edit_distance
from nltk.util import ngrams
nltk.download('words')
nltk.download('stopwords')
nltk.download('punkt')
import re

def clean_text(text):
  # remove new lines and tab spaces
  text = text.replace('\n',' ')
  text = text.replace('\t',' ')
  # remove mentions and urls
  text = re.sub(r"(?:\@|https?\://)\S+", "", text)
  text = text.lower()
  # remove punctuation and emojis
  text = re.sub(r'[^\w\s#]', '', text)
  emoji_pattern = re.compile("["
          u"\U0001F600-\U0001F64F"  # emoticons
          u"\U0001F300-\U0001F5FF"  # symbols & pictographs
          u"\U0001F680-\U0001F6FF"  # transport & map symbols
          u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            "]+", flags=re.UNICODE)
  text = text.encode('utf-8', errors='ignore').decode('utf-8') # remove non utf-8
  text = re.sub(r'\d+', '', text) # remove numbers
  text = emoji_pattern.sub(r'', text) # no emoji
  return text


from gensim.parsing.preprocessing import remove_stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk_stopwords = set(stopwords.words('english'))
def remove_stop_words(text):
  # remove stop words
  text = remove_stopwords(text) #gensim method
  text_tokens = word_tokenize(text)
  text_tokens = [word for word in text_tokens if not word in nltk_stopwords] #nltk method
  text = ' '.join(text_tokens)
  return text

from textblob import TextBlob
# word correction
def word_correction(text):
  return TextBlob(text).correct()

# lemmatization
import spacy
nlp = spacy.load("en_core_web_sm")
def lemmatize(text):        
    sent = []
    doc = nlp(text)
    for word in doc:
        if len(word) < 3:
          continue
        if '@' in word.text:  # these are hashtags and should be left unaltered
          sent.append(word.text)
        else:
          sent.append(word.lemma_)
    text =  " ".join(sent)
    return text.replace('-PRON-', '')

```

```python
sample_text = '@Sam #web3 #FTX #FTT #binance #Metaverse #Airdrop #tokens #moon #airdrops #airdrop #freeairdrop #getfree #free https://t.co/UfHNIYobpj caltural'

cleaned_text = clean_text(sample_text)
cleaned_text = lemmatize(cleaned_text.replace('#','@')) #converting # to @ for lemmatizer to leave the hashtags unaltered
cleaned_text = remove_stop_words(cleaned_text)
cleaned_text = cleaned_text.replace('@','') #removing the @ symbols and the hashtag terms are retained as individual words
cleaned_text = re.sub(' +', ' ', cleaned_text)
cleaned_text = word_correction(cleaned_text)
print(cleaned_text)
```

```python
from tqdm import tqdm
clean_tweets = []

n = len(df.text)
for i in tqdm(range(n)):
    cleaned_text = clean_text(df.text.iloc[i])
    cleaned_text = lemmatize(cleaned_text.replace('#','@')) #converting # to @ for lemmatizer to leave the hashtags unaltered
    cleaned_text = remove_stop_words(cleaned_text)
    cleaned_text = cleaned_text.replace('@','') #removing the @ symbols and the hashtag terms are retained as individual words
    cleaned_text = re.sub(' +', ' ', cleaned_text)
    # cleaned_text = word_correction(cleaned_text)
    clean_tweets.append(cleaned_text)
clean_tweets[:5]
```
